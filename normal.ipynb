```python
import numpy as np
from scipy.special import erfinv
import matplotlib.pyplot as plt

# 正規分布 (-1,1)の一様分布から正規分布を得るための関数
def normal_dist(p):
    return np.sqrt(2) * erfinv(p)

# 一様分布からサンプル生成
u = np.random.uniform(-1, 1, size=100000)

# プロットして確認
plt.hist(normal_dist(u), bins=1000, density=True, alpha=0.7, label='Transformed')
x = np.linspace(-4, 4, 20000)

# 本当の正規分布をプロット
plt.plot(x, 1/np.sqrt(2*np.pi) * np.exp(-x**2 / 2), 'r--', label='N(0,1) PDF')
plt.legend()
plt.title("Transforming Uniform(-1,1) to N(0,1)")
plt.show()
```
