# filtration of Vietoris-Rips Complexes

## 例1

```python
import gudhi
import numpy as np
import matplotlib.pyplot as plt

# 2次元のサンプル点群（円形）
theta = np.linspace(0, 2 * np.pi, 30+1)
circle1 = np.vstack((np.cos(theta), np.sin(theta))).T
points = np.vstack([circle1])

# Rips複体の構築
rips_complex = gudhi.RipsComplex(points=points, max_edge_length=20.0)
simplex_tree = rips_complex.create_simplex_tree(max_dimension=2)

# パーシステンス図を計算
simplex_tree.compute_persistence()

# === プロット部分 ===
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 1. 点群の可視化
axes[0].scatter(points[:, 0], points[:, 1], c="blue")
axes[0].set_title("data sets")
axes[0].axis("equal")

# 2. パーシステンス図
gudhi.plot_persistence_diagram(simplex_tree.persistence(), axes=axes[1])
axes[1].set_title("persistence disgram")

plt.tight_layout()
plt.show()

# 3. 1次ホモロジーのみ可視化せずに数字として出力
intervals_by_dim = {}
for p in simplex_tree.persistence():
    if p[0] == 1:
        print(p)
```

```python
import gudhi
import numpy as np
import matplotlib.pyplot as plt

# 2次元のサンプル点群（円形）
theta = np.linspace(0, 2 * np.pi, 30+1)
circle1 = np.vstack((2 + 2*np.cos(theta), 2*np.sin(theta))).T
circle2 = np.vstack((-1 + np.cos(theta), np.sin(theta))).T
points = np.vstack([circle1, circle2])

# Rips複体の構築
rips_complex = gudhi.RipsComplex(points=points, max_edge_length=10)
simplex_tree = rips_complex.create_simplex_tree(max_dimension=2)

# パーシステンス図を計算
simplex_tree.compute_persistence()

# === プロット部分 ===
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 1. 点群の可視化
axes[0].scatter(points[:, 0], points[:, 1], c="blue")
axes[0].set_title("data sets")
axes[0].axis("equal")

# 2. パーシステンス図
gudhi.plot_persistence_diagram(simplex_tree.persistence(), axes=axes[1])
axes[1].set_title("persistence disgram")

plt.tight_layout()
plt.show()

# 3. 1次ホモロジーのみ可視化せずに数字として出力
intervals_by_dim = {}
for p in simplex_tree.persistence():
    if p[0] == 1:
        print(p)
```
